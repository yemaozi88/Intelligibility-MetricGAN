```python
import os
import sys
import glob
import librosa
import numpy as np
import soundfile as sf

repos_dir = r'/home/takkan/repos'
sys.path.append(repos_dir)
from sak import display as dp
from sak import signal_processing as sp
from sak import nele

from pesq import pesq
```

```python
wavDir_in  = r'/home/takkan/experiments/test_pesq_audio'

input_wav_paths = glob.glob(os.path.join(wavDir_in, '*.wav'))
input_wav_paths.sort()
print(len(input_wav_paths))
for i in range(len(input_wav_paths)):
    print(i, input_wav_paths[i])
```

```python
print(input_wav_paths[0])
sample_signal_clean, sample_sr_clean = sf.read(input_wav_paths[0])
print(len(sample_signal_clean),sample_signal_clean, sample_sr_clean)
print(input_wav_paths[1])
sample_signal_noise, sample_sr_noise = sf.read(input_wav_paths[1])
print(len(sample_signal_noise), sample_signal_noise, sample_sr_noise)
```

```python
print(pesq(sample_sr_clean, sample_signal_clean, sample_signal_noise, 'wb'))
print(pesq(sample_sr_clean, sample_signal_clean, sample_signal_noise, 'nb'))
```

```python
print(input_wav_paths[2])
#signal_clean = sp.load_wav(input_wav_paths[2])
signal_clean, signal_sr_clean = librosa.load(input_wav_paths[2], sr=16000)
print(len(signal_clean),signal_clean,signal_sr_clean)

print(input_wav_paths[5])
#signal_snr_10 = sp.load_wav(input_wav_paths[5])
signal_snr_10, signal_sr_snr_10 = librosa.load(input_wav_paths[5], sr=16000)
print(len(signal_snr_10),signal_snr_10,signal_sr_snr_10)
```
